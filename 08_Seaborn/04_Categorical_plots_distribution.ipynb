```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

### BoxPlot

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df, hue = 'gender',width=0.4)
```

```python
plt.figure(figsize=(12,20))
sns.boxplot(y='parental level of education',x='math score',data=df, hue = 'race/ethnicity', orient='h')
```

### Violin Plot

```python
plt.figure(figsize=(12,6))
sns.violinplot(y='parental level of education',x='math score',data=df, hue='gender',split=True, inner='quartile')
```

### Swarm Plot

```python
sns.swarmplot(x='math score',data=df, size = 4)
```

```python
plt.figure(figsize=(12,10))
sns.swarmplot(x='math score',y = 'race/ethnicity',data=df, size = 4)
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot(y='math score',x = 'race/ethnicity',data=df, size = 4, hue='gender')
```
