```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
#rug plot
sns.rugplot(data=df,x='salary',height = 0.5)
```

```python
#displot : show you the distribution of a single feature. It is a histogram with the option of adding 'KDE' plot
sns.displot(data=df, x= 'sales',kde = True)
```

```python
#histogram
sns.set(style = 'white')
sns.histplot(data=df, x='salary',bins=10, color = 'blue', edgecolor='red')
plt.savefig('demo.png')
```

```python
import matplotlib.pyplot as plt

# Data
x = [1] * 10
y = [10, 20, 30, 40, 50, 60, 70, 80, 90, 1000]

from sklearn.preprocessing import MinMaxScaler

# Define the MinMaxScaler
scaler = MinMaxScaler()

# Reshape the data to a 2D array and fit the scaler
y_reshaped = np.array(y).reshape(-1, 1)
scaler.fit(y_reshaped)

# Transform the data
y_scaled = scaler.transform(y_reshaped)

# Reshape the scaled data back to a 1D array
y = y_scaled.flatten()

# Create the plot
fig, ax = plt.subplots()

# Plot data
ax.plot(x, y, 'o', color='white')

# Set axes and labels color to white
ax.spines['top'].set_color('white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Set the background to transparent
fig.patch.set_alpha(0)
ax.patch.set_alpha(0)

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

plt.savefig('demo1.png')
# Show plot
plt.show()
```
