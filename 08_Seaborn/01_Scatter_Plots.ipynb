```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

#### Scatterplot

```python
sns.scatterplot(data=df,x='salary',y='sales')
plt.title('this is title')
plt.show()
```

```python
#hue : color points based off a categorical feature in the DF
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='salary',y='sales',hue='division')
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='salary',y='sales',hue='work experience', palette = 'inferno')
```

```python
df['training level'].unique()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=df,x='salary',y='sales',hue='work experience', palette = 'inferno', size = 'training level')
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data=df,
                x='salary',
                y='sales',
                hue='work experience', 
                palette = 'inferno', 
                size = 'training level',
               style = 'division')
plt.savefig('example_scatter.png')
```
