```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

### Catplot()

```python
plt.figure(figsize=(10,30))
sns.catplot(x='gender',y='reading score',data=df,kind='violin',
            row='race/ethnicity', col= 'test preparation course',
            hue='lunch')
plt.subplots_adjust(hspace=0.2,wspace=1)
```

### Facetgrid

```python
g = sns.FacetGrid(data=df, col='gender',row='lunch')
g.map(plt.scatter,'math score','reading score')
```
