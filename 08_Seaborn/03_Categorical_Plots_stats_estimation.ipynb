```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

#### Countplot()

```python
plt.figure(figsize=(12,4))
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(12,4))
sns.countplot(data=df,x='division',hue='level of education',palette='Set3')
```

### Barplot()

```python
plt.figure(figsize=(12,4))
sns.barplot(x='level of education', y='sales',data=df, estimator=np.max, hue = 'division')
plt.ylabel('max sales')
plt.legend(loc=(1.02,0.5))
```
