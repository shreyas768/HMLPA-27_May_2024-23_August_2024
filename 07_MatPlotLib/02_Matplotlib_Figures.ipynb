### Add multiple set of axes to the figure

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) #Large Figure
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) #Small Figure
axes3 = fig.add_axes([1.2,0.5,.4,.4])

axes1.plot(a,b)
axes1.set_xlabel('this is a label')
axes1.set_ylabel('this is b label')
axes1.set_title('Big Figure')

axes2.plot(c,d)
axes2.set_xlabel('this is c label')
axes2.set_xlabel('this is d label')
axes2.set_xlim(2,8)
axes2.set_ylim(2.5,15)
axes2.set_title('Zoomed In Small Figure')

axes3.plot(d,c)
axes3.set_title('d vs c')


plt.savefig('demo.png')
plt.show()
```
