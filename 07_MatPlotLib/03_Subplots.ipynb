```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b=a**3
```

```python
c = np.arange(0,10)
d = 3*c
```

#### Subplots

```python
fig, axes = plt.subplots()

axes.plot(c,d,'red')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('this is title')  #hide out []
```

```python
fig, axes = plt.subplots(nrows=4,ncols=3)
```

```python
axes.shape
```

```python
#axes here is an array of axes to plot on
axes
```

### plotting on axes object

```python
fig, axes = plt.subplots(nrows=2,ncols=2)

for p in axes:
    p[0].plot(a,b)
    p[1].plot(c,d)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(a,b)
axes[-1].plot(c,d)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2, figsize = (12,7))
axes[1,1].plot(a,b)
axes[1,1].set_title('this is title')

axes[0,1].plot(c,d)
axes[0,1].set_xlabel('this is xlabel for 0,1')

axes[1,0].plot(d,c)
axes[0,0].plot(b,a)

fig.subplots_adjust(wspace=0.9,hspace=0.9)

fig.suptitle('this is title for entire image')
fig.savefig('demo.png')
```
