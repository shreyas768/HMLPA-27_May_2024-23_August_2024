```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

#### Basic Matplotlib Commands

```python
plt.plot(x,y)
plt.xlabel('X axis label here')
plt.ylabel('Y axis label here')
plt.title('Plot Title here')
plt.show()  #required for non-jupyter/colab users
```

#### Editing more figure parameters

```python
plt.plot(x,y)
plt.xlabel('X axis label here')
plt.ylabel('Y axis label here')
plt.title('Plot Title here')
plt.xlim(1,6) #lower limit, upper limit
plt.ylim(2.5,10)
plt.savefig('demo.png')  #Exporting the image
plt.show()  #required for non-jupyter/colab users
```
