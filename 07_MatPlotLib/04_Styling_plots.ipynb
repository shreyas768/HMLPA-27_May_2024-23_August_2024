```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0,10,11)
b=a**3
c = np.arange(0,10)
d = 3*c
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(a,a**2, label = 'a**2')
ax.plot(a,a**4, label = 'a**4')
ax.legend(loc=(1.2,0.5))

#1 = upper right
#2 = upper left
#3 = lower left
#4 = #lower right
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(a,a**2,color = 'green', label = 'a**2', ls = '--',lw = 2.10, alpha = 0.3)
ax.plot(a,a**4, color = '#CCE5FF', label = 'a**4') #RBG hex code
ax.plot(a, a**3, color='blue', marker='P', linestyle=':',linewidth=2, markersize=15) #, mfc = 'red',mew = 9)
ax.legend(loc=(1.2,0.5))
```
