## Predicting Labels

```python
def display(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data, x='X1',y='X2',hue=labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans
km2_model = KMeans(n_clusters=2)
```

```python
display(km2_model,moons)
```

```python
km3_model = KMeans(n_clusters=3)
display(km3_model,blobs)
```

```python
display(km2_model,circles)
```

### DBSCAN Results

```python
from sklearn.cluster import DBSCAN
```

```python
db1 = DBSCAN(eps=0.8)
```

```python
display(db1,blobs)
```

```python
db2 = DBSCAN(eps=0.2)
display(db2,moons)
```
