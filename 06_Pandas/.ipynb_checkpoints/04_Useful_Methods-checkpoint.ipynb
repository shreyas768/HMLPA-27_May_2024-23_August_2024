```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
def last_six(n):
    return int(str(n)[-6:])
```

```python
last_six(237218947874902901)
```

```python
df['last_six_CC'] = df['CC Number'].apply(last_six)
df.head()
```

```python
df['total_bill'].mean()
```

```python
def range1(price):
    if price < 10:
        return '*'
    elif price >=10 and price < 20:
        return '**'
    else:
        return '***'
```

```python
df['bill_category'] = df['total_bill'].apply(range1)
df.head()
```

```python
df['CC Number'].apply(lambda n : int(str(n)[-6:]))
```

* tip/total_bill < 12 -->> thanks
* 12 >= tip/total_bill < 15 -->> kind
* 15 >= tip/total_bill < 19 -->> generous
* tip/total_bill > 19  -->> political

df['tip_category']

```python
def ratio1(tip,total_bill):
    if (tip/total_bill) < 0.12:
        return 'thanks'
    elif 0.12 <= (tip/total_bill) < .15:
        return 'kind'
    elif 0.15 <= (tip/total_bill) <.19:
        return 'generous'
    else:
        return 'political'
    
```

```python
df['tip_quality'] = df[['tip','total_bill']].apply(lambda df : ratio1(df['tip'],df['total_bill']),axis=1)
```
